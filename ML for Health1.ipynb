# using icu table - urine

```python

query_icu_urine = f"""
SELECT
  le.subject_id,
  le.hadm_id,
  dlab.label,
  le.value AS urine_output,
  le.charttime,
  AVG(ie.patientweight) AS mean_weight_kg
FROM
  `physionet-data.mimiciv_icu.outputevents` AS le
JOIN
  `physionet-data.mimiciv_icu.d_items` AS dlab ON le.itemid = dlab.itemid
JOIN
  `physionet-data.mimiciv_icu.inputevents` AS ie ON le.subject_id = ie.subject_id AND le.hadm_id = ie.hadm_id
WHERE
  dlab.label LIKE "Foley%"
GROUP BY
  le.subject_id, le.hadm_id, dlab.label,le.value,le.charttime
ORDER BY
  le.subject_id
  """

query_job = client.query(query_icu_urine)
urine_icu_df = query_job.to_dataframe()

print(urine_icu_df)

```

```python

merged_icu_df = pd.merge(urine_icu_df, unique_patients, on=['subject_id', 'hadm_id'], how='left', indicator=True)

hypertensive_urine_icu_patients = merged_icu_df[merged_icu_df['_merge'] == 'both']

print(hypertensive_urine_icu_patients[['subject_id', 'hadm_id', 'urine_output', 'label', 'charttime', 'mean_weight_kg']])
print(len(hypertensive_urine_icu_patients))
```

```python


hypertensive_urine_icu_patients['urine_per_kg_hr'] = (
    hypertensive_urine_icu_patients['urine_output'] / hypertensive_urine_icu_patients['mean_weight_kg']
)

hypertensive_urine_icu_patients.sort_values(by=['subject_id', 'charttime'], inplace=True)

def find_consecutive_low_urine(group):
    results = []
    low_urine_start = None

    for index, row in group.iterrows():
        if row['urine_per_kg_hr'] < 0.5:

            if low_urine_start is None:
                low_urine_start = row['charttime']

            duration = row['charttime'] - low_urine_start

            if duration >= pd.Timedelta(hours=6):
                results.append(row.to_dict())
        else:

            low_urine_start = None

    return pd.DataFrame(results)

consecutive_low_urine_icu_df = hypertensive_urine_icu_patients.groupby(['subject_id', 'hadm_id']).apply(find_consecutive_low_urine)

consecutive_low_urine_icu_df.reset_index(drop=True, inplace=True)

print(consecutive_low_urine_icu_df)
```

```python

unique_subject_ids_urine = consecutive_low_urine_icu_df['subject_id'].unique()
print("\nUnique subject_ids with consecutive low urine output:")
print(unique_subject_ids_urine)
print(len(unique_subject_ids_urine))
```

## **Statistical Analysis (if this part is not useful, maybe can delete)**

**1 Normality test**: Kolmogorov-Smirnov test




Out_join_df Columns:

* subject_id: patients unique identifier.
* hadm_id: Admission unique identifier.
* label_x: It comes from the urine measurement data, that is, the label column in the consecutive_low_urine_icu_df dataset.
* label_y: It comes from the creatinine measurement data, i.e., the label column in the found_subset dataset.
* urine_output: Measured value of urine output.
* charttime_x and charttime_y: measurement timestamp, corresponding to urine and creatinine data respectively.
* mean_weight_kg: The average of patient weight (used to normalize urine output).
* urine_per_kg_hr: Urine output per kilogram of body weight per hour, which is an early indicator related to AKI.
* value: Measured value of serum creatinine.
* Creatinine_Difference: The difference in creatinine value changes within 48 hours, used to evaluate significant changes in renal function.

```python

import pandas as pd
from scipy.stats import kstest

# Kolmogorov-Smirnov Test
continuous_columns = ['urine_per_kg_hr', 'Creatinine_Difference', 'mean_weight_kg']

for column in continuous_columns:
    stat, p_value = kstest(outer_join_df[column], 'norm')
    if p_value > 0.05:
        print(f"{column}: may conform to normal distribution (p-value = {p_value:.3f})")
    else:
        print(f"{column}: may not  conform to normal distribution (p-value = {p_value:.3f})")
```

  * Although urine_output and value are also continuous data, they can better reflect individual differences after standardization (such as converting urine_output into urine output per kilogram of body weight per hour). Therefore, they are included in the urine_per_kg_hr variable and no longer undergo a separate normality test.
  * charttime_x and charttime_y are timestamps and are not suitable for normality tests because they are not continuous variables representing physiological measurements.
  * **urine_per_kg_hr**, **Creatinine_Difference**, **mean_weight_kg**: The p-values of the three variables are all 0.000, which means that these variables do not conform to the normal distribution. According to the Kolmogorov-Smirnov test, if the p-value is less than 0.05, we can conclude that the data does not conform to the normal distribution. Therefore, **these variables may have skewness or extreme values**.

**2 Descriptive statistics**: Normally distributed variables can be expressed as mean  standard deviation, and non-normally distributed variables can be expressed as median and interquartile range (IQR). Categorical variables can be expressed as frequency and percentage.


```python
# Descriptive statistics for categorical variables
categorical_columns = ['label_x', 'label_y']

for column in categorical_columns:
    counts = outer_join_df[column].value_counts()
    percentages = outer_join_df[column].value_counts(normalize=True) * 100
    print(f"\n{column} Categorical Variable Statistics:\n", pd.DataFrame({'Count': counts, 'Percentage': percentages}))

# Group comparison
from scipy.stats import ttest_ind, mannwhitneyu, chi2_contingency, fisher_exact

# Assuming 'Creatinine_Difference' is used to classify patients into AKI or non-AKI groups
aki_group = outer_join_df[outer_join_df['Creatinine_Difference'] >= 0.3]
non_aki_group = outer_join_df[outer_join_df['Creatinine_Difference'] < 0.3]
```

This shows that in the dataset, almost all patients use the same urine measurement method, which is good for data consistency, but also means that there may be a lack of diversity.

```python
# Descriptive statistics function

def descriptive_statistics(df, column):
    if kstest(df[column], 'norm').pvalue > 0.05:
        # Normal distribution
        mean = df[column].mean()
        std = df[column].std()
        print(f"{column} descriptive statistics: Mean = {mean:.2f}, Standard Deviation = {std:.2f}")
    else:
        # Non-normal distribution
        median = df[column].median()
        iqr = df[column].quantile(0.75) - df[column].quantile(0.25)
        print(f"{column} descriptive statistics: Median = {median:.2f}, Interquartile Range (IQR) = {iqr:.2f}")

# Descriptive statistics for continuous variables
for column in continuous_columns:
    descriptive_statistics(outer_join_df, column)

```

**3 Group comparison**: Students t test or Mann-Whitney U test was used to compare continuous variables, and chi-square test or Fishers exact test was used to compare categorical variables.

**The Mann-Whitney U test**, suitable for comparing non-normally distributed continuous variables, showed **significant differences** in urine output, creatinine changes, and body weight between the AKI and non-AKI groups.

```python
# Group comparison for continuous variables
for column in continuous_columns:
    if kstest(outer_join_df[column], 'norm').pvalue > 0.05:
        # If normal distribution, use Student's t-test
        stat, p_value = ttest_ind(aki_group[column], non_aki_group[column])
        print(f"Student's t-test for {column}: p-value = {p_value:.3f}")
    else:
        # If non-normal distribution, use Mann-Whitney U test
        stat, p_value = mannwhitneyu(aki_group[column], non_aki_group[column])
        print(f"Mann-Whitney U test for {column}: p-value = {p_value:.3f}")
```

**Fisher's exact test**, suitable for categorical variables with small sample sizes or low counts, showed a **significant association** between measurement type and the occurrence of AKI.

```python
# Group comparison for categorical variables
for column in categorical_columns:
    contingency_table = pd.crosstab(outer_join_df['Creatinine_Difference'] >= 0.3, outer_join_df[column])
    if contingency_table.min().min() < 5:
        # Use Fisher's exact test
        _, p_value = fisher_exact(contingency_table)
        print(f"Fisher's exact test for {column}: p-value = {p_value:.3f}")
    else:
        # Use Chi-squared test
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        print(f"Chi-squared test for {column}: p-value = {p:.3f}")
```

```python
#  Sample the datasets
sampled_dfs = {}
for name, df in dfs.items():
    # For large datasets, use a smaller fraction
    if len(df) > 500000:
        sample_frac = 0.01  # Sample 1% of larger datasets
    else:
        sample_frac = 0.05  # Sample 5% of smaller datasets
    sampled_df = df.sample(frac=sample_frac, random_state=42)
    sampled_dfs[name] = sampled_df
    print(f"{name}: Sampled {int(sample_frac * 100)}%, new shape {sampled_df.shape}")
```

```python
# Balance the labels (AKI vs non-AKI)

# Combine AKI and non-AKI patients
aki_df = sampled_dfs['outer_join_df']
non_aki_df = sampled_dfs['non_aki_patients_df']

# Resample non-AKI patients to match the number of AKI patients
if len(aki_df) > len(non_aki_df):
    aki_downsampled = resample(aki_df, replace=False, n_samples=len(non_aki_df), random_state=42)
    balanced_df = pd.concat([aki_downsampled, non_aki_df], axis=0)
else:
    non_aki_downsampled = resample(non_aki_df, replace=True, n_samples=len(aki_df), random_state=42)
    balanced_df = pd.concat([aki_df, non_aki_downsampled], axis=0)

balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)  # Shuffle the balanced dataset
print(f"Balanced dataset shape: {balanced_df.shape}")

# Update the sampled_dfs dictionary with the balanced labels
balanced_labels = balanced_df[['subject_id', 'hadm_id', 'AKI_label']]
sampled_dfs['balanced_labels'] = balanced_labels

# Save the balanced dataset for further usage
balanced_labels.to_csv('balanced_labels.csv', index=False)

# Print final shape of all sampled datasets
for name, df in sampled_dfs.items():
    print(f"{name}: Final shape {df.shape}")
```

# RESULTS/ANALYSIS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score, accuracy_score, precision_score, recall_score, f1_score
from collections import defaultdict
from prettytable import PrettyTable
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.calibration import calibration_curve
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression, Lasso
from sklearn.model_selection import StratifiedKFold, train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
```

```python
class AnalysisRunner():
    def __init__(self, data):        
        # Define global PrettyTables
        self.table_balance = PrettyTable()
        self.table_balance.field_names = [
            "Fold", 
            "(T) 0%", 
            "(T) 0 Count", 
            "(T) 1%", 
            "(T) 1 Count", 
            "(V) 0%", 
            "(V) 0 Count", 
            "(V) 1%", 
            "(V) 1 Count", 
            "Train/Val Instances"
        ]

        self.table_results = PrettyTable()
        self.table_results.field_names = [
            "Fold", 
            "Model", 
            "Validation Accuracy", 
            "Validation Precision", 
            "Validation Recall", 
            "Validation F1 Score"
        ]

        # Prepare features and target variable
        self.y = data['AKI_label']  # Label
        self.x = data.drop(columns=['AKI_label'])  # Features
        self.x = self.x.drop(columns=['subject_id', 'hadm_id'], errors='ignore')
        # Select features and target variable
        self.y = pd.Series(np.where(self.y == -1.0, 0, self.y), index=self.y.index)

        # Data Partition
        self.xT, self.xV, self.yT, self.yV = train_test_split(self.x, self.y, test_size=0.1, random_state=42)
        # Store metrics for aggregation
        self.model_metrics = defaultdict(lambda: {"accuracy": [], "precision": [], "recall": [], "f1": [], "roc_auc": []})

        # For ROC calculation
        self.models = {
            "SVM": SVC(kernel='rbf', random_state=42, probability=True),  # probability=True for ROC
            "XGBoost": XGBClassifier(eval_metric='logloss', random_state=42),
            "Decision Tree": DecisionTreeClassifier(random_state=42),
            "Naive Bayes": GaussianNB(),
            "Random Forest": RandomForestClassifier(random_state=42),
            "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42)
        }

    # Add the ROC curve method
    def plotRocCurves(self, n_splits=10):
        # Define StratifiedKFold
        skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

        plt.figure(figsize=(12, 10))

        # Iterate through each model
        for model_name, model in self.models.items():
            tprs, fprs = [], []
            aucs = []
            # Perform K-Fold cross-validation
            for fold_idx, (train_index, val_index) in enumerate(skf.split(self.x, self.y), start=1):
                x_train, x_val = self.x.iloc[train_index], self.x.iloc[val_index]
                y_train, y_val = self.y.iloc[train_index], self.y.iloc[val_index]                
                # Train the model
                model.fit(x_train, y_train)                
                # Predict probabilities (for ROC curve)
                if hasattr(model, "predict_proba"):
                    y_val_proba = model.predict_proba(x_val)[:, 1]  # Probability of the positive class
                else:
                    y_val_proba = model.decision_function(x_val)  # For models like SVM                
                # Compute ROC curve
                fpr, tpr, _ = roc_curve(y_val, y_val_proba)
                roc_auc = roc_auc_score(y_val, y_val_proba)
                # Append TPR, FPR, and AUC
                tprs.append(tpr)
                fprs.append(fpr)
                aucs.append(roc_auc)
            # Compute mean AUC for this model
            mean_auc = np.mean(aucs)
            self.model_metrics[model_name]["roc_auc"].append(mean_auc)
            # Plot the ROC curve for the model
            plt.plot(fprs[-1], tprs[-1], label=f"{model_name} (AUC = {mean_auc:.5f})")

        # Plot formatting
        plt.plot([0, 1], [0, 1], linestyle='--', color='gray')  # Diagonal
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curves for Each Model')
        plt.legend(loc="lower right")
        plt.tight_layout()
        plt.show()

        # Function to generate calibration curves for models
    def plotCalibrationCurves(self, n_splits=10):
        # Define StratifiedKFold
        skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

        plt.figure(figsize=(12, 8))

        # Iterate through each model
        for model_name, model in self.models.items():
            true_prob_list = []
            pred_prob_list = []

            # Perform K-Fold cross-validation
            for fold_idx, (train_index, val_index) in enumerate(skf.split(self.x, self.y), start=1):
                x_train, x_val = self.x.iloc[train_index], self.x.iloc[val_index]
                y_train, y_val = self.y.iloc[train_index], self.y.iloc[val_index]
                
                # Train the model
                model.fit(x_train, y_train)
                
                # Predict probabilities (for calibration curve)
                if hasattr(model, "predict_proba"):
                    y_val_proba = model.predict_proba(x_val)[:, 1]  # Probability of the positive class
                else:
                    y_val_proba = model.decision_function(x_val)  # For models like SVM

                # Append true labels and predicted probabilities
                true_prob_list.append(y_val)
                pred_prob_list.append(y_val_proba)
            
            # Flatten the lists
            true_prob_list = np.concatenate(true_prob_list)
            pred_prob_list = np.concatenate(pred_prob_list)

            # Generate calibration curve
            prob_true, prob_pred = calibration_curve(true_prob_list, pred_prob_list, n_bins=10)
            
            # Plot the calibration curve for this model
            plt.plot(prob_pred, prob_true, marker='o', label=f"{model_name}")

        # Plot formatting
        plt.plot([0, 1], [0, 1], linestyle='--', color='gray')  # Perfectly calibrated line
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('Predicted Probability')
        plt.ylabel('True Probability')
        plt.title('Calibration Curves for Each Model')
        plt.legend(loc="best")
        plt.tight_layout()
        plt.show()

    def learnBestFeatures(self):
        # Feature standardization
        xTscaled = StandardScaler().fit_transform(self.xT)
        # LASSO Regression
        lasso = Lasso(alpha=0.01)
        lasso_coefficients = lasso.fit(xTscaled, self.yT).coef_
        # Create a DataFrame to display feature names and their corresponding coefficients
        feature_importance = pd.DataFrame({
            'Feature': self.x.columns,
            'Coefficient': lasso_coefficients
        })
        # Filter out features with non-zero coefficients
        selected_features = feature_importance[feature_importance['Coefficient'] != 0]
        # Sort by the absolute value of the coefficients in descending order
        selected_features = selected_features.sort_values(by='Coefficient', key=abs, ascending=False)
        return selected_features
    
    def describeFeaturesImportance(self):
        param_grid = {
            'n_estimators': [100, 200, 300],
            'max_depth': [3, 5, 7],
            'learning_rate': [0.01, 0.1, 0.2],
            'subsample': [0.8, 1.0]
        }
        grid_search = GridSearchCV(XGBClassifier(eval_metric='logloss'), param_grid, cv=5, scoring='accuracy')
        grid_search.fit(self.xT, self.yT)    
        best_params = grid_search.best_params_
        model = XGBClassifier(**best_params, eval_metric='logloss')
        model.fit(self.xT, self.yT)    
        explainer = shap.Explainer(model)
        shap_values = explainer(self.xT)
        shap.summary_plot(shap_values, self.xT)

    def doValidation(self, data, selected_feature_names, n_splits=10):
        def evaluate_data_balance(x_train, y_train, x_val, y_val, fold_idx):
            # Training set statistics
            train_samples, n_features = x_train.shape
            unique_labels_train, counts_train = np.unique(y_train, return_counts=True)
            count_0_train = counts_train[unique_labels_train.tolist().index(0)] if 0 in unique_labels_train else 0
            count_1_train = counts_train[unique_labels_train.tolist().index(1)] if 1 in unique_labels_train else 0
            percentage_0_train = (count_0_train / train_samples) * 100
            percentage_1_train = (count_1_train / train_samples) * 100
            
            # Validation set statistics
            val_samples = x_val.shape[0]
            unique_labels_val, counts_val = np.unique(y_val, return_counts=True)
            count_0_val = counts_val[unique_labels_val.tolist().index(0)] if 0 in unique_labels_val else 0
            count_1_val = counts_val[unique_labels_val.tolist().index(1)] if 1 in unique_labels_val else 0
            percentage_0_val = (count_0_val / val_samples) * 100
            percentage_1_val = (count_1_val / val_samples) * 100
            
            # Add data to the global balance table in a single row
            self.table_balance.add_row([
                f"Fold {fold_idx}",
                f"{percentage_0_train:.3f}%", 
                count_0_train,
                f"{percentage_1_train:.3f}%",
                count_1_train,
                f"{percentage_0_val:.3f}%",
                count_0_val,
                f"{percentage_1_val:.3f}%",
                count_1_val,
                f"{train_samples}/{val_samples}"
                
            ])
        # Define StratifiedKFold
        skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

        # Iterate through each fold
        for fold_idx, (train_index, val_index) in enumerate(skf.split(self.x, self.y), start=1):
            x_train, x_val = self.x.iloc[train_index], self.x.iloc[val_index]
            y_train, y_val = self.y.iloc[train_index], self.y.iloc[val_index]
            evaluate_data_balance(x_train, y_train, x_val, y_val, fold_idx)
            
            # Train, evaluate, and store results for each model
            for model_name, model in self.models.items():
                model.fit(x_train, y_train)
                y_val_pred = model.predict(x_val)
                accuracy = accuracy_score(y_val, y_val_pred)
                precision = precision_score(y_val, y_val_pred, zero_division=1)
                recall = recall_score(y_val, y_val_pred, zero_division=1)
                f1 = f1_score(y_val, y_val_pred, zero_division=1)
                self.model_metrics[model_name]["accuracy"].append(accuracy)
                self.model_metrics[model_name]["precision"].append(precision)
                self.model_metrics[model_name]["recall"].append(recall)
                self.model_metrics[model_name]["f1"].append(f1)
                self.table_results.add_row([
                    f"Fold {fold_idx}",
                    model_name,
                    f"{accuracy:.4f}",
                    f"{precision:.4f}",
                    f"{recall:.4f}",
                    f"{f1:.4f}"
                ])

    # Function to aggregate and print the model results
    def aggregate_model_results(self):
        # Define a PrettyTable for the aggregated results
        table_aggregated = PrettyTable()
        table_aggregated.field_names = [
            "Model", 
            "Avg Accuracy", 
            "Avg Precision", 
            "Avg Recall", 
            "Avg F1 Score", 
            "Avg ROC AUC"
        ]
        # Calculate and add the averages for each model
        for model_name, metrics in self.model_metrics.items():
            avg_accuracy = np.mean(metrics["accuracy"])
            avg_precision = np.mean(metrics["precision"])
            avg_recall = np.mean(metrics["recall"])
            avg_f1 = np.mean(metrics["f1"])
            avg_roc_auc = np.mean(metrics["roc_auc"]) if "roc_auc" in metrics else "N/A"
            
            # Add aggregated metrics to the aggregated results table
            table_aggregated.add_row([
                model_name,
                f"{avg_accuracy:.4f}",
                f"{avg_precision:.4f}",
                f"{avg_recall:.4f}",
                f"{avg_f1:.4f}",
                f"{avg_roc_auc:.4f}"
            ])
        return table_aggregated
```
